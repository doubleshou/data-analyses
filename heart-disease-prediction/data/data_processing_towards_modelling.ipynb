```python
import pandas as pd
```

```python
READ_PATH = "../data/heart_2020_cleaned.csv"
SAVE_PATH = "../data/heart_2020_for_modelling.pkl"
```

```python
heart = pd.read_csv(READ_PATH)
```

```python
heart.head()
```

```python
heart["HeartDisease"] = heart["HeartDisease"].replace({"No": 0, "Yes": 1})
```

```python
heart_dumm = pd.get_dummies(heart)
```

```python
heart_dumm.head()
```

```python
heart_dumm.info()
```

```python
heart_dumm.to_pickle(SAVE_PATH)
```
